```python
import pandas as pb
import numpy as np
from matplotlib import pyplot as plt
from sklearn.preprocessing import MinMaxScaler
df=pb.read_csv("customer_churn_large_dataset.csv",encoding="utf-8")

"""Checking of Null spaces """
df.isnull().sum()


"""Feature Engineering
performing normalization on the data set"""

df.replace(to_replace=["Male","Female"],value=[1,0],inplace=True)
df[["Monthly_Bill"]]=MinMaxScaler().fit_transform(df[["Monthly_Bill"]])
df[["Subscription_Length_Months"]]=MinMaxScaler().fit_transform(df[["Subscription_Length_Months"]])
df[["Total_Usage_GB"]]=MinMaxScaler().fit_transform(df[["Total_Usage_GB"]])

"""Separating the data for training """
x=df[["Gender","Subscription_Length_Months","Monthly_Bill","Total_Usage_GB"]]
y=df[["Churn"]]

"""Splitting the Data to be trained and data to be tested  """
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.9,random_state=57)




```

```python
"""Building the neural network for training the model """
def neural():

    from tensorflow import keras
    from keras import Sequential
    modle=Sequential(
        [keras.layers.Dense(20,input_shape=(4,),activation="relu"),
         keras.layers.Dense(10,activation="relu"),
        keras.layers.Dense(1,activation="sigmoid")]
    )
    modle.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
    return modle

```

```python
""" Performing cross validation to check which model is the best  """
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,f1_score,precision_score,recall_score,roc_auc_score

"""Loading the models in a dictionary  """

modles={"DecisionTreeClassifier" :DecisionTreeClassifier(),
        "RandomForestClassifier" :RandomForestClassifier(),
        "LogisticRegression":LogisticRegression()}


"""Loading the models one by one and training them"""
for i in range(0,len(modles)):
    modle=list(modles.values())[i]
    modle.fit(xtrain,ytrain)
    ytrain_pred=modle.predict(xtrain)
    ytest_pred=modle.predict(xtest) 
    
    
    """Checking the performance of the model by using performance metrics   """
    
    
    print("model name : ",list(modles.keys())[i])
    print("Train score")
    print("accuracy_score : ",accuracy_score(ytrain_pred,ytrain))
    print("f1_score : ",f1_score(ytrain_pred,ytrain))
    print("precision_score : ",precision_score(ytrain_pred,ytrain))
    print("recall_score : ",recall_score(ytrain_pred,ytrain))
    print("roc_auc_score : ",roc_auc_score(ytrain_pred,ytrain))
    print("___________________________________")
    print("___________________________________")
    print("test score")
    print("accuracy_score : ",accuracy_score(ytest_pred,ytest))
    print("f1_score : ",f1_score(ytest_pred,ytest))
    print("precision_score : ",precision_score(ytest_pred,ytest))
    print("recall_score : ",recall_score(ytest_pred,ytest))
    print("roc_auc_score : ",roc_auc_score(ytest_pred,ytest))
    print("___________________________________")
    print("___________________________________")

    
```

```python
"""loading the neural network model and training it """
modle=neural()
modle.fit(xtrain,ytrain,epochs=50)
ytrain_value=modle.predict(xtrain)
ytest_value=modle.predict(xtest)

ytrain_pred=[]
ytest_pred=[]

"""converting the prediction to 0s and 1s"""
for i in ytrain_value:
    if i>0.5:
        ytrain_pred.append(1)
    else:
        ytrain_pred.append(0)
for i in ytest_value:
    if i>0.5:
        ytest_pred.append(1)
    else:
        ytest_pred.append(0)

"""Checking the performance of the neural network"""
print("model name : Neural network")
print("Train score")
print("accuracy_score : ",accuracy_score(ytrain_pred,ytrain))
print("f1_score : ",f1_score(ytrain_pred,ytrain))
print("precision_score : ",precision_score(ytrain_pred,ytrain))
print("recall_score : ",recall_score(ytrain_pred,ytrain))
print("roc_auc_score : ",roc_auc_score(ytrain_pred,ytrain))
print("___________________________________")
print("___________________________________")
print("test score")
print("accuracy_score : ",accuracy_score(ytest_pred,ytest))
print("f1_score : ",f1_score(ytest_pred,ytest))
print("precision_score : ",precision_score(ytest_pred,ytest))
print("recall_score : ",recall_score(ytest_pred,ytest))
print("roc_auc_score : ",roc_auc_score(ytest_pred,ytest))
print("___________________________________")
print("___________________________________")
```
